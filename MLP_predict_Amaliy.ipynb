**Overfitting (Ortiqcha moslashish):** Bu model o'quv ma'lumotlariga (training data) juda yaxshi moslashib, hatto shovqinni ham o'rganib olganda yuz beradi. Natijada, yangi, ko'rilmagan ma'lumotlar (test data) ustida ish samaradorligi yomon bo'ladi. Oddiy qilib aytganda, model 'yodlab olgan' bo'ladi, lekin tushunmagan.

**Underfitting (Kam moslashish):** Bu model o'quv ma'lumotlaridagi asosiy tendensiyalarni o'rgana olmaganda sodir bo'ladi. Model juda sodda bo'ladi va ma'lumotlardagi muhim naqshlarni ilg'ay olmaydi. Natijada, ham o'quv, ham yangi ma'lumotlar ustida ish samaradorligi yomon bo'ladi. Oddiy qilib aytganda, model yetarlicha o'rganmagan bo'ladi.

**Bias (Xatolik):** Modelning noto'g'ri taxminlar tufayli ma'lumotlardagi haqiqiy munosabatni o'rganishda xatolikka yo'l qo'yishi. Underfitting ko'pincha yuqori bias bilan bog'liq.

**Variance (Tarqalish):** Modelning turli xil o'quv ma'lumotlariga qanchalik sezgir ekanligini bildiradi. Yuqori variance model turli ma'lumotlar to'plamlarida juda boshqacha xatti-harakat qilishini anglatadi. Overfitting ko'pincha yuqori variance bilan bog'liq.

Maqsad har doim bias va variance o'rtasida optimal muvozanatni topish, shunda model ham o'quv ma'lumotlariga yaxshi moslashadi, ham yangi ma'lumotlarda umumlashtira oladi.



//////////////////////////

**Parametrlar (Parameters):**

Nima bu? Parametrlar modelning o'zi o'quv ma'lumotlaridan (training data) o'rganadigan ichki o'zgaruvchilardir. Boshqacha qilib aytganda, ular modelning "o'zidan-o'zi" aniqlanadigan qismlaridir.

Qanday aniqlanadi? Modelni o'qitish (training) jarayonida, ma'lumotlar orqali optimallashtirish algoritmlari (masalan, gradient descent) yordamida bu parametrlar sozlanadi va ularning qiymatlari yangilanadi.

Misollar: Neyron tarmoqlaridagi vaznlar (weights) va ofsetlar (biases), chiziqli regressiyadagi koeffitsiyentlar (coefficients).

Maqsad: Parametrlar modelning ma'lumotlardagi naqshlarni topishi va bashorat qilish qobiliyatini belgilaydi.

**Giperparametrlar (Hyperparameters):**

Nima bu? Giperparametrlar modelning tuzilishi yoki o'qitish jarayonini boshqaradigan tashqi konfiguratsiya sozlamalaridir. Ular modelning o'zi emas, balki modelning qanday o'rganishini nazorat qiladigan sozlamalar.

Qanday aniqlanadi? Giperparametrlar modelni o'qitishdan oldin inson tomonidan (yoki avtomatik qidiruv algoritmlari, masalan, Grid Search, Random Search yordamida) belgilanadi va ular o'quv ma'lumotlaridan to'g'ridan-to'g'ri o'rganilmaydi.

Misollar: Neyron tarmog'idagi qatlamlar soni (number of layers), har bir qatlamdagi neyronlar soni (number of neurons per layer), o'rganish tezligi (learning rate), muntazamlik parametri (regularization parameter), K-eng yaqin qo'shnilardagi K qiymati (K in K-Nearest Neighbors).

Maqsad: Giperparametrlar modelning ishlashini optimallashtirishga yordam beradi, masalan, overfitting yoki underfittingdan qochish, modelning tezroq va samaraliroq o'rganishini ta'minlash.
Asosiy farq:

Parametrlar modelning ichki qismidir va ma'lumotlardan o'rganiladi.
Giperparametrlar modelni o'qitishdan oldin o'rnatiladigan tashqi konfiguratsiyalardir va ular ma'lumotlardan o'rganilmaydi, balki modelning qanday o'rganishini nazorat qiladi.

Optimal giperparametrlar modelning eng yaxshi ishlashiga erishish uchun juda muhimdir.

##**sun'iy neyron tarmoqlarning (SNT) eng keng tarqalgan turlari**

**To'liq bog'langan neyron tarmoqlar (Feedforward Neural Networks - FNN / Multilayer Perceptron - MLP):**

Nima bu? Eng sodda va asosiy SNT turi. Ma'lumotlar faqat bir yo'nalishda, kirishdan chiqishga qarab harakatlanadi. Har bir neyron oldingi qatlamdagi barcha neyronlarga to'liq bog'langan bo'ladi. Ular ma'lumotlar orasidagi murakkab, chiziqsiz bog'liqliklarni o'rganishga qodir. Bir yoki bir nechta yashirin qatlamlarga ega bo'lishi mumkin.

Qo'llanilishi: Umuman olganda, tasniflash (classification), regressiya (regression) kabi har qanday jadval ma'lumotlari yoki oddiyroq naqshlarni aniqlashda ishlatiladi.

Shu vaqatgacha depp learningda o'tgan darslar bu MLP ga misol bo'ladi.

**Konvolyutsion neyron tarmoqlar (Convolutional Neural Networks - CNN):**

Nima bu? Asosan tasvirlarni va shunga o'xshash tuzilmadagi ma'lumotlarni (masalan, audio) qayta ishlash uchun maxsus mo'ljallangan. Ular konvolyutsiya, pooling va to'liq bog'langan qatlamlardan iborat. Konvolyutsiya qatlamlari kichik filtrlar (kernel) yordamida tasvirning turli qismlaridan o'tib, lokal xususiyatlarni (chegaralar, burchaklar, ranglar) aniqlaydi va o'rganadi. Bu tarmoqlarning tasvirdagi shift invariantligini (joylashuvga nisbatan befarqlik) ta'minlaydi.

Qo'llanilishi: Tasvirlarni tanib olish (object recognition), yuzni aniqlash, tibbiy tasvirlarni tahlil qilish va kompyuter ko'rishi bilan bog'liq boshqa vazifalar.

**Rekurrent neyron tarmoqlar (Recurrent Neural Networks - RNN):**

Ketma-ket ma'lumotlar (matn, nutq, vaqt qatorlari) bilan ishlash uchun mo'ljallangan. Ichki xotiraga ega, bu ularga oldingi bosqichlardagi ma'lumotlarni eslab qolishga imkon beradi.

**Uzoq qisqa muddatli xotira tarmoqlari (LSTM) va Gated Recurrent Units (GRU):**

 RNNlarning takomillashtirilgan versiyalari bo'lib, ular uzoq muddatli bog'liqliklarni yaxshiroq o'rganadi. Gate (darvoza) mexanizmlari orqali ma'lumotni eslab qolish yoki unutishni boshqaradi.

**Avtoenkoderlar (Autoencoders):**

 Ma'lumotlarni siqish (encoding) va keyin qayta tiklash (decoding) orqali ma'lumotlardagi muhim xususiyatlarni o'rganadi. Ko'pincha o'lchamni kamaytirish va shovqinni tozalash uchun ishlatiladi.

**Generativ raqib tarmoqlar (Generative Adversarial Networks - GANs):**

 Haqiqiyga o'xshash yangi ma'lumotlar (masalan, tasvirlar) yaratish uchun bir-biri bilan raqobatlashadigan ikkita tarmoqdan (generator va diskriminator) iborat.

**Transformerlar (Transformers):**

 E'tibor mexanizmiga (attention mechanism) asoslangan. Asosan tabiiy tilni qayta ishlashda (NLP) qo'llaniladi va uzoq bog'liqliklarni juda samarali hal qiladi.

# **Deep Learning**

## Xususiyatlar va Maqsadni Ajratish va Tensorlarga O'tkazish

### Subtask:
Modelga kiritish uchun `Churn_prepared` DataFrame'idan xususiyatlar (`X_tensor`) va maqsad (`y_tensor`) o'zgaruvchilarini ajratish va ularni PyTorch tensorlariga aylantirish. Bunda `y_tensor` `df['Exited']` ustunidan olinadi.


```python
import torch

# 1. `Churn_prepared` DataFrame'idan 'Exited' ustunini tashlab, qolgan ustunlarni xususiyatlar (X) sifatida ajrating.
X = Churn_prepared.drop('Exited', axis=1)

# 2. Original `df` DataFrame'idan 'Exited' ustunini maqsad (y) sifatida ajrating.
y = df['Exited']

# 3. `X` va `y` ni NumPy massivlariga aylantiring.
X_np = X.values
y_np = y.values

# 4. `torch` kutubxonasini import qiling. (Already imported above)

# 5. `X` NumPy massivini `torch.tensor()` yordamida `torch.float32` turidagi `X_tensor`ga aylantiring.
X_tensor = torch.tensor(X_np, dtype=torch.float32)

# 6. `y` NumPy massivini `torch.tensor()` yordamida `torch.float32` turidagi `y_tensor`ga aylantiring
# va uni ustun vektoriga aylantirish uchun `.reshape(-1, 1)` dan foydalaning.
y_tensor = torch.tensor(y_np, dtype=torch.float32).reshape(-1, 1)

print("X_tensor type:", X_tensor.dtype)
print("X_tensor shape:", X_tensor.shape)
print("y_tensor type:", y_tensor.dtype)
print("y_tensor shape:", y_tensor.shape)
```

## Ma'lumotlarni O'qitish va Test To'plamlariga Ajratish

### Subtask:
Tayyorlangan tensor ma'lumotlarini `train_test_split` yordamida o'qitish va test to'plamlariga (`X_train`, `X_test`, `y_train`, `y_test`) ajratish.


```python
from sklearn.model_selection import train_test_split

# Ma'lumotlarni o'qitish va test to'plamlariga ajratish
X_train, X_test, y_train, y_test = train_test_split(
    X_tensor, y_tensor, test_size=0.2, random_state=42, stratify=y_tensor
)

# Hosil bo'lgan tensorlar shakllarini chop etish
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

## Optimal MLP Modelini Belgilash

### Subtask:
Eng yuqori aniqlikni bergan MLP model arxitekturasini (`nn.Dropout` qatlamlarisiz) `nn.Sequential` yordamida qayta aniqlash.


```python
import torch.nn as nn

# Kirish va chiqish o'lchamlarini aniqlash
input_size = X_train.shape[1]
output_size = y_train.shape[1]

# MLP modelini yaratish
model = nn.Sequential(
    nn.Linear(input_size, 64),  # Birinchi yashirin qatlam
    nn.ReLU(),
    nn.Linear(64, 32),          # Ikkinchi yashirin qatlam
    nn.ReLU(),
    nn.Linear(32, output_size), # Chiqish qatlami
    nn.Sigmoid()                # Binary classification uchun Sigmoid
)

# Model arxitekturasini chop etish
print(model)
```

## Yo'qotish Funksiyasi va Optimizatorni Belgilash

### Subtask:
Ikkilik tasniflash uchun `nn.BCELoss` yo'qotish funksiyasini va eng yaxshi natija bergan `torch.optim.SGD` optimizatorini `learning_rate=0.01` bilan belgilash.


```python
import torch.nn as nn
import torch.optim as optim

# 1. nn.BCELoss yo'qotish funksiyasini criterion o'zgaruvchisiga belgilang.
criterion = nn.BCELoss()

# 2. torch.optim.SGD optimizatorini optimizer o'zgaruvchisiga belgilang, unga modelning o'rganiladigan parametrlarini (model.parameters()) va learning_rate=0.01 ni kiriting.
optimizer = optim.SGD(model.parameters(), lr=0.01)

print("Loss function (criterion) defined as:", criterion)
print("Optimizer defined as:", optimizer)
```

## MLP Modelini O'qitish

### Subtask:
Belgilangan MLP modelini o'qitish ma'lumotlarida (`X_train`, `y_train`) o'qitish jarayonini amalga oshirish va o'qitish davomidagi yo'qotish (loss) grafigini chizish.


```python
import matplotlib.pyplot as plt

# 1. O'qitish davomiyligini (epochs) 100 ga o'rnating.
epochs = 100

# 2. Yo'qotish qiymatlarini saqlash uchun bo'sh ro'yxat (`losses`) yarating.
losses = []

# 3. Har bir epoch uchun o'qitish siklini bajaring:
for epoch in range(epochs):
    # a. Modelni o'qitish rejimiga o'tkazing (`model.train()`).
    model.train()

    # b. Oldingi gradientlarni tozalang (`optimizer.zero_grad()`).
    optimizer.zero_grad()

    # c. Model orqali `X_train` ma'lumotlarini o'tkazib, bashoratlarni (`outputs`) oling.
    outputs = model(X_train)

    # d. Bashoratlar va `y_train` yordamida yo'qotish (`loss`) hisoblang.
    loss = criterion(outputs, y_train)

    # e. Yo'qotish qiymati bo'yicha orqaga tarqalishni bajaring (`loss.backward()`).
    loss.backward()

    # f. Optimizator qadamini bajaring (`optimizer.step()`).
    optimizer.step()

    # g. Hisoblangan yo'qotish qiymatini `losses` ro'yxatiga qo'shing (`loss.item()`).
    losses.append(loss.item())

    # h. Har 10 epochda o'qitish yo'qotishini chop eting.
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}')

# 4. `matplotlib.pyplot` yordamida `losses` ro'yxatidagi yo'qotish qiymatlarini aks ettiruvchi grafik chizing.
plt.figure(figsize=(10, 6))
plt.plot(losses)
plt.title("MLP Modelining O'qitish Yo'qotilishi")
plt.xlabel('Epoch')
plt.ylabel('Yo\'qotish')
plt.grid(True)
plt.show()
```

## MLP Modelini Baholash

### Subtask:
O'qitilgan modelning test to'plamidagi (`X_test`, `y_test`) aniqligini hisoblash va natijani chiqarish.


```python
import torch

# 1. Modelni baholash rejimiga o'tkazing.
model.eval()

# 2. `torch.no_grad()` kontekst menejeridan foydalanib, `X_test` ma'lumotlarida bashoratlarni (`outputs`) oling.
with torch.no_grad():
    outputs = model(X_test)

# 3. Bashoratlarni 0.5 chegarasiga qarab ikkilik qiymatlarga aylantiring.
predicted = (outputs > 0.5).float()

# 4. Bashorat qilingan qiymatlarni (`predicted`) haqiqiy qiymatlar (`y_test`) bilan solishtirib, aniqlikni hisoblang.
accuracy = (predicted == y_test).float().mean() * 100

# 5. Hisoblangan aniqlikni chop eting.
print(f'Test aniqligi: {accuracy.item():.2f}%')
```
